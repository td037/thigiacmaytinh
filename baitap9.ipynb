```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_path = 'D:/Thi giac may tinh/image/6.jpg'
my_img = Image.open(my_path)
plt.imshow(my_img)
```

```python
im_gray = my_img.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
plt.figure()
plt.hist(im_array.flatten(), bins=128)
plt.title('Bieu do histogram' )
plt.xlabel('gia tri cua pixel')
plt.ylabel('so luong')
plt.show()
```

```python
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```
